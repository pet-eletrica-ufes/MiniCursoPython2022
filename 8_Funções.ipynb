# Definição


> Para simplificar o nosso código principal, podemos fazer uso de Funções. 

> Basicamente são uma sequência de comandos que realizam determinada tarefa e recebem um nome.

> Formato Geral:
  
    def nome (argumento, argumento, ...):
      comandos
      ...
      comandos

````python
def Fibonacci (n):
  # Escreve a série de Fibonacci até o número mais próximo e menor que n:
  a,b = 0,1
  print (a)
  while (b<n):
    print (b)
    a,b = b, a+b 
````

````python
Fibonacci (50)
````

> Existe um comando chamado **Return** que é utilizado para retornar o resultado da função.

> Ele finaliza a função e retoma a execução do código principal no ponto onde a função foi chamada


````python
def soma (x,y):
  resultado = x + y
  return resultado
````

````python
x = int(input("Valor de X: ")) 
y = int(input("Valor de Y: "))

resp = soma (x, y)

print("O valor retornado foi: ", resp)
````

# Argumentos


> Se uma função possui X argumentos, sua chamada deve possui X valores para ser compatível. 

> Caso especial: Argumentos **Default**

````python
def func(nome, saudacao="Bom dia", pontuacao="!"):
    return saudacao+nome+pontuacao
````

````python
print (func(" Maria"))
````

````python
print (func(" Maria", "Boa tarde"))
````

````python
print (func(" Maria", "Boa noite", "!!!!" ))
````

````python
print (func(saudacao = "Tchau", nome=" Maria"))
````

# Documentando Funções


> Poderíamos utilizar comentários para descreverem o objetivo da função que estamos desenvolvendo. Porém usar **docstrings** é mais vantajoso.

> Basicamente escreve-se uma string após o cabeçalho da função.

> O comando `"funçao".__doc__ ` retorna o conteúdo dessa string.

````python
def fatorial(n):
  "Retorna o fatorial de n."
  for i in range (n-1, 1,-1):
    n*=i
  return n
````

````python
fatorial(5)
````

````python
print (fatorial.__doc__)
````

# Funções Especiais

> Notação *Lambda* 

>> É possível definir uma função simples (que possui uma única expressão) em apenas uma linha por meio da Notação *Lambda*

>> Formato:

  ```
            # lambda <parametros>: <codigo de retorno>
  ```





````python
f = lambda a,b : a+b
````

````python
f(10,2)
````
